```python
# Example feature: Average runs scored by a team in the last 15 games
def calculate_avg_runs_last15(data, team_id_col, runs_col):
    data['avg_runs_last15'] = data.groupby(team_id_col)[runs_col].transform(lambda x: x.rolling(15, 1).mean())
    return data

train_data = calculate_avg_runs_last15(train_data, 'team1_id', 'inning1_runs')
train_data = calculate_avg_runs_last15(train_data, 'team2_id', 'inning2_runs')

test_data = calculate_avg_runs_last15(test_data, 'team1_id', 'inning1_runs')
test_data = calculate_avg_runs_last15(test_data, 'team2_id', 'inning2_runs')
```
