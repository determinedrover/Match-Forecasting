```python
import pandas as pd
import numpy as np
from tqdm import tqdm
tqdm.pandas()
```

```python
data_bowlers = pd.read_csv('bowler_level_scorecard.csv')
```

```python
data_bowlers.head()
```

```python
print(len(data_bowlers['bowler_id'].unique()))
```

```python
bowler_summary = data_bowlers.groupby('bowler_id').agg({
    'runs': 'sum',
    'wicket_count': 'sum',
    'balls_bowled': 'sum',
    'economy': 'mean',
    'maiden': 'sum',
    'dots': 'sum',
    'Fours': 'sum',
    'Sixes': 'sum',
    'wides': 'sum',
    'noballs': 'sum'
}).reset_index()
```

```python
output_file_path = 'summarized_bowler_data.csv'
bowler_summary.to_csv(output_file_path, index=False)
```

```python
def calculate_bowler_rating(row):
    # Parameters for weighting the factors
    weights = {
        'runs': -0.01,
        'wicket_count': 3.0,
        'balls_bowled': 0.1,
        'economy': -1.0,
        'maiden': 0.5,
        'dots': 0.1,
        'Fours': -0.2,
        'Sixes': -0.3,
        'wides': -0.5,
        'noballs': -0.5
    }
    
    # Calculate the rating
    rating = (
        row['runs'] * weights['runs'] +
        row['wicket_count'] * weights['wicket_count'] +
        row['balls_bowled'] * weights['balls_bowled'] +
        row['economy'] * weights['economy'] +
        row['maiden'] * weights['maiden'] +
        row['dots'] * weights['dots'] +
        row['Fours'] * weights['Fours'] +
        row['Sixes'] * weights['Sixes'] +
        row['wides'] * weights['wides'] +
        row['noballs'] * weights['noballs']
    )
    

    
    return rating
```

```python
summarized_bowler_df = pd.read_csv('summarized_bowler_data.csv')

summarized_bowler_df['rating'] = summarized_bowler_df.apply(calculate_bowler_rating, axis=1)
```
